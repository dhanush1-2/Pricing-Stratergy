```python
import pandas as pd

df = pd.read_csv("data/retail_sales_dataset.csv")
df.head()
```

```python
df = df[["Total Amount", "Price per Unit"]].rename(columns={"Total Amount": "Sales", "Price per Unit": "Price"})
df.head()
```

```python
df.shape
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 5))
plt.scatter(df["Price"], df["Sales"], alpha=0.5)
plt.xlabel("Price")
plt.ylabel("Sales")
plt.title("Price vs Sales")
plt.show()
```
